```python
#Importing packages
import random 
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#Patient Record list to store the dictionaries of patient information
patient_record = []
```

```python
#Function for Enter Patient Information
def enter_patient_info():
    print("Enter Patient Information:")
    initial_service_date = input("Date of Service (MMDDYYYY): ")
    patient_id = int(input("Patient ID#: "))
    name = input("Name: ").upper()
    sex = input("Sex (M/F): ").upper()
    age = int(input("Age: "))
    race = input("Race/Ethnicity: ").upper()
    prev_opioid_type = input("Previous Opioid Use Type: ").upper()
    initial_methadone_dose = int(input("Methadone Dose (mg): "))
    
    #Adding inputs to dictionary
    patient_info = {
        "Patient ID": patient_id,
        "Name": name,
        "Sex": sex,
        "Age": age,
        "Race/Ethnicity": race,
        "Previous Opioid Use Type": prev_opioid_type,
        "Date of Service": initial_service_date,
        "Methadone Dose (mg)": initial_methadone_dose,
    }
    patient_record.append(patient_info)
    print("Patient {} information is added/updated successfully.".format(name))  #need to put before return in order to print
    return
    
```

```python
#Function to append patient's methadone dose and date of service
def update_methadone_dose():

    #asks user for input 
    patient_id = input("Enter Patient ID: ")
    new_date = input("Enter Date of Service (MMDDYYYY): ")
    new_dose = input("Enter Methadone Dose (mg): ")
    
    #makes sure dose is written as an integer
    try:
        new_dose = int(new_dose)
    except ValueError:
        print("Invalid dose entered. Please enter a numeric value.")
        return
    
    #finds patient record
    patient_found = False
    for record in patient_record:
        if record["Patient ID"] == patient_id:
            patient_found = True
            
            #'date of service' and 'methadone dose' are lists
            if not isinstance(record["Date of Service"], list):
                record["Date of Service"] = [record["Date of Service"]]
            if not isinstance(record["Methadone Dose (mg)"], list):
                record["Methadone Dose (mg)"] = [record["Methadone Dose (mg)"]]
            
            #to append new date and dose
            record["Date of Service"].append(new_date)
            record["Methadone Dose (mg)"].append(new_dose)
            print(f"Patient ID {patient_id} updated successfully.")
            break
    
    if not patient_found:
        print(f"Patient ID {patient_id} not found.")
```

```python
#Look up patient record
def lookup():
    patient_id = input("Enter Patient ID: ")
    patient_found = False
    for record in patient_record: 
        if record ["Patient ID"] == patient_id:
            print(record)
            patient_found = True
            break
    if not patient_found: 
        print("Patient not found.")
```

```python
#Function for the input prompt/welcome screen
def intro_prompt():
    while True:
        print("Welcome!")
        print("1. Enter Patient Information")
        print("2. Update Patient's Methadone Dose")
        print("3. Look Up Patient Information")
        print("4. Exit")

        choice = input("Please choose an option: ")

        if choice == "1":
            enter_patient_info()
        elif choice == "2":
            update_methadone_dose()
        elif choice == "3":
            lookup()
        elif choice == "4":
            print("Goodbye.")
            break
        else:
            print("Please enter a valid option.")
            break
intro_prompt()
```

```python
#to calculate retention rate of patients 
""" retention rate = (number of patients retained at end of period 
divided by number of patients at start of period) x 100 """

#convert 'date_of_service' to date time format (needed by python)
df['date_of_service'] = pd.to_datetime(df['date_of_service'], format='%mm-%dd-%yyyy')

#filter date_of_service for month of november
november_data = df[(df['date_of_service'] >= '11-01-2024') & (df['date_of_service'] <= '11-30-2024')]

#remove duplicates in 'patient_id' 
total_patients = len(set(df['patient_id']))

#filter patients by removing patients who received 0 mg into a new list
valid_patients = []
for patient_id, group in november_data.groupby('patient_id'):
    if group['methadone_dose_mg'].min() > 0:  
        valid_patients.append(patient_id)

#makes sure valid_patients list is in filtered dataframe by using .isin
valid_patients = november_data[november_data['patient_id'].isin(valid_patients)]

#filtered retained patients into a new list
retained_patients = [] #patients who received a dose from day 1 to day 30
for patient_id, group in valid_patients.groupby('patient_id'):
    start_date = group['date_of_service'].min()
    end_date = group['date_of_service'].max()
    if start_date <= pd.Timestamp('11-01-2024') and end_date >= pd.Timestamp('11-30-2024'):
        retained_patients.append(patient_id)
        
#calculate retention rate
retained_count = len(retained_patients)
retention_rate = (retained_count / total_patients) * 100
retention_rate
```

```python
#calculate average methadone dose per patient
average_dose_list = []
patients_total = list(set(df['patient_id']))

for patient_id in patients_total:
    # Filter data for the current patient
    patient_data = df[df['patient_id'] == patient_id]['methadone_dose_mg']
    #calculate average dose for patient
    average_dose = np.mean(patient_data)
    #append to list
    average_dose_list.append({'patient_id': patient_id, 'average_methadone_dose_mg': average_dose})

#convert list to df
average_dose_per_patient = pd.DataFrame(average_dose_list)
average_dose_per_patient
```
